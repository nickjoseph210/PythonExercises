```python
iris = sns.load_dataset('iris')

iris.dtypes
```

```python
df=iris
sns.distplot(df.petal_length)
```

```python
# is there a correlation between petal_length and petal_width?

sns.relplot(x='petal_length', y='petal_width', data=iris)


# Yes.  Directly proportional.
```

```python
# reasonable to make a prediction based on sepal width and sepal length?

sns.relplot(x='sepal_width', y='sepal_length', data=iris)

# Nope.  Too many data points on either side of the vector arrow.
```

```python
# Which features would be best to predict species? 

sns.relplot(x='petal_length', y='sepal_length', data=iris)

sns.pairplot(iris)

# Looks like it would be sepal length and petal length
```
